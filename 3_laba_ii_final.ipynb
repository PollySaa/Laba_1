# **1. Импорт данных.**

```python
import pandas as pd

data = pd.read_csv('NY-House-Dataset.csv', sep=',')
data
```

# **2. Определение задачи**

```python
percentile_1 = data['PRICE'].quantile(0.01)
percentile_99 = data['PRICE'].quantile(0.99)
filtered_data = data[(data['PRICE'] > percentile_1) & (data['PRICE'] < percentile_99)]
filtered_data["PRICE"].hist()
```

# **3. Обработка данных**

**A**

```python
from sklearn.model_selection import train_test_split

X = filtered_data.drop('PRICE', axis=1)
y = filtered_data['PRICE']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=17)
X_train
```

**B**

```python
filtered_data.info()
```

**C**

```python
import numpy as np

y_train_log = np.log(y_train)
y_test_log = np.log(y_test)

y_test_log
```

**D**

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np

categorical_columns = ['BROKERTITLE', 'TYPE', 'ADDRESS', 'STATE', 'MAIN_ADDRESS', 'ADMINISTRATIVE_AREA_LEVEL_2', 'LOCALITY', 'SUBLOCALITY', 'STREET_NAME', 'LONG_NAME', 'FORMATTED_ADDRESS',]

ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_train_encoded = ohe.fit_transform(X_train[categorical_columns])
X_test_encoded = ohe.transform(X_test[categorical_columns])

encoded_feature_names = ohe.get_feature_names_out(categorical_columns)

X_train_encoded_d = pd.DataFrame(X_train_encoded, columns=encoded_feature_names, index=X_train.index)
X_test_encoded_d = pd.DataFrame(X_test_encoded, columns=encoded_feature_names, index=X_test.index)

X_test_encoded
```

# **4. Обучаем на тренировочном множестве**

**A**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train_encoded, y_train_log)
lr_test_pred = lr.predict(X_test_encoded)
lr_train_pred = lr.predict(X_train_encoded)
```

**B**

```python
from sklearn.tree import DecisionTreeRegressor

max_depths = range(1, 11)
trees = {}
tree_y_pred = {}

for depth in max_depths:
    tree = DecisionTreeRegressor(max_depth=depth, random_state=17)
    tree.fit(X_train_encoded, y_train_log)

    trees[depth] = tree
    tree_y_pred[depth] = tree.predict(X_test_encoded)
```

**C**

```python
from sklearn.neighbors import KNeighborsRegressor

max_neighbors = range(1, 11)
knns = {}
knn_y_pred = {}

for neighbor in max_neighbors:
    knn = KNeighborsRegressor(n_neighbors=neighbor)
    knn.fit(X_train_encoded, y_train_log)

    knns[neighbor] = tree
    knn_y_pred[neighbor] = knn.predict(X_test_encoded)
```

**D**

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=50, max_depth=5, random_state=17)
rfr.fit(X_train_encoded, y_train_log)
rfr_y_pred = rfr.predict(X_test_encoded)
```

# **5. Считаем метрики на train и test множествах**

**A**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

y_test_exp = np.exp(y_test_log)
y_train_exp = np.exp(y_train_log)
lr_test_pred_exp = np.exp(lr_test_pred)
lr_train_pred_exp = np.exp(lr_train_pred)

mask = y_train_exp != 0

mae_test = mean_absolute_error(y_test_exp, lr_test_pred_exp)
rmse_test = np.sqrt(mean_squared_error(y_test_exp, lr_test_pred_exp))
mape_test = np.mean(np.abs((y_test_exp - lr_test_pred_exp) / y_test_exp)) * 100

mae_train = mean_absolute_error(y_train_exp, lr_train_pred_exp)
rmse_train = np.sqrt(mean_squared_error(y_train_exp, lr_train_pred_exp))
mape_train = np.mean(np.abs((y_train_exp[mask] - lr_train_pred_exp[mask]) / y_train_exp[mask])) * 100

print(f"MAE Test = {mae_test}")
print(f"MAE Train = {mae_train}")
print(f"RMSE Test = {rmse_test}")
print(f"RMSE Train = {rmse_train}")
print(f"MAPE Test = {mape_test}")
print(f"MAPE Train = {mape_train}")
```

**B**

```python
for depth in max_depths:
    tree_test_perd_exp = np.exp(tree_y_pred[depth])
    tree_train_pred_exp = np.exp(trees[depth].predict(X_train_encoded))

    mae_test = mean_absolute_error(y_test_exp, tree_test_perd_exp)
    rmse_test = np.sqrt(mean_squared_error(y_test_exp, tree_y_pred[depth]))
    mape_test = np.mean(np.abs((y_test_exp - tree_test_perd_exp) / y_test_exp)) * 100

    mae_train = mean_absolute_error(y_train_exp, tree_train_pred_exp)
    rmse_train = np.sqrt(mean_squared_error(y_train_exp, tree_train_pred_exp))
    mape_train = np.mean(np.abs((y_train_exp[mask] - tree_train_pred_exp[mask]) / y_train_exp[mask])) * 100

    print(f"Depth = {depth}\n")
    print(f"MAE Test = {mae_test}")
    print(f"MAE Train = {mae_train}")
    print(f"RMSE Test = {rmse_test}")
    print(f"RMSE Train = {rmse_train}")
    print(f"MAPE Test = {mape_test}")
    print(f"MAPE Train = {mape_train}\n\n")
```

```python
for neighbor in max_neighbors:
    neighbor_test_pred = np.exp(knn_y_pred[neighbor])
    neighbor_train_pred = np.exp(knns[neighbor].predict(X_train_encoded))

    mae_test = mean_absolute_error(y_test_exp, neighbor_test_pred)
    rmse_test = np.sqrt(mean_squared_error(y_test_exp, neighbor_test_pred))
    mape_test = np.mean(np.abs((y_test_exp - neighbor_test_pred) / y_test_exp)) * 100

    mae_train = mean_absolute_error(y_train_exp, neighbor_train_pred)
    rmse_train = np.sqrt(mean_squared_error(y_train_exp, neighbor_train_pred))
    mape_train = np.mean(np.abs((y_train_exp[mask] - neighbor_train_pred[mask]) / y_train_exp[mask])) * 100

    print(f"Neighbors = {neighbor}\n")
    print(f"MAE Test = {mae_test}")
    print(f"MAE Train = {mae_train}")
    print(f"RMSE Test = {rmse_test}")
    print(f"RMSE Train = {rmse_train}")
    print(f"MAPE Test = {mape_test}")
    print(f"MAPE Train = {mape_train}\n\n")
```
