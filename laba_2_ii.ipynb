0. Импорта данных

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('NY-House-Dataset.csv', sep=',')
data
```

```python
percentile_1 = data['PRICE'].quantile(0.01)
percentile_99 = data['PRICE'].quantile(0.99)
filtered_data = data[(data['PRICE'] > percentile_1) & (data['PRICE'] < percentile_99)]
filtered_data
```

2. Описание постановки задачи



Цель: Предсказать цену на недвижимость в Нью-Йорке (целевая переменная - PRICE).

Тип задачи: Регрессия (предсказываем число, так как цена — непрерывная переменная).

Переменные:

PRICE: Целевая переменная, цена на недвижимость.

BEDROOMS: Количество спален.

BATH: Количество ванных комнат.

PROPERTYSQFT: Площадь недвижимости в квадратных футах.

TYPE: Тип недвижимости (например, Condo, House, Townhouse, Co-op).

ADDRESS: Адрес недвижимости.

STATE: Штат (в данном случае, все записи относятся к Нью-Йорку).

MAIN_ADDRESS: Основной адрес.

ADMINISTRATIVE_AREA_LEVEL_2: Административный уровень 2 (например, New York County).

LOCALITY: Местность (например, New York).

SUBLOCALITY: Подместность (например, Manhattan).

STREET_NAME: Название улицы.

LONG_NAME: Длинное название.

FORMATTED_ADDRESS: Форматированный адрес.

LATITUDE: Широта.

LONGITUDE: Долгота.

Целевое событие
Целевое событие — это PRICE, которое является непрерывной переменной. Мы будем предсказывать число, представляющее цену на недвижимость.

8. Графики зависимости от целевой переменной

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='BEDS', y='PRICE', data=filtered_data)
plt.title('Зависимость цены от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Цена')
plt.show()
```

Вывод: если на графике видно, что с увеличением количества спален цена увеличивается, то это указывает на положительную корреляцию между BEDS и PRICE. Это значит, что количество спален может быть полезным признаком для прогнозирования цены.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='BATH', y='PRICE', data=filtered_data)
plt.title('Зависимость цены от количества ванн')
plt.xlabel('Количество ванн')
plt.ylabel('Цена')
plt.show()
```

Вывод: если с увеличением количества ванн цена увеличивается, то это также указывает на положительную корреляцию между BATH и PRICE. Количество ванн может быть полезным признаком для прогнозирования цены.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='PROPERTYSQFT', y='PRICE', data=filtered_data)
plt.title('Зависимость цены от площади недвижемости')
plt.xlabel('Количество площади недвижемости')
plt.ylabel('Цена')
plt.show()
```

Вывод: если с увеличением площади недвижемости цена увеличивается, то это указывает на положительную корреляцию между PROPERTYSQFT и PRICE. Площадь недвижемости может быть важным признаком для прогнозирования цены.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='LATITUDE', y='PRICE', data=filtered_data)
plt.title('Зависимость цены от широты')
plt.xlabel('Количество широты')
plt.ylabel('Цена')
plt.show()
```

Вывод: если на графике видно, что цена меняется в зависимости от широты, то это указывает на наличие корреляции между LATITUDE и PRICE. Широта может быть полезным признаком для прогнозирования цены, особенно если учитывать, что стоимость недвижимости может зависеть от местоположения (например, близость к центру города или другим важным объектам).

Заключение: если на всех графиках наблюдается какая-либо зависимость (линейная или нелинейная) между переменными BEDS, BATH, PROPERTYSQFT, LATITUDE и целевой переменной PRICE, то эти переменные могут быть использованы для прогнозирования цены. Чем сильнее зависимость, тем более информативными будут эти переменные для модели прогнозирования.
