```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np

categorical_columns = ['TYPE', 'STATE', 'LOCALITY', 'SUBLOCALITY', 'STREET_NAME', 'LONG_NAME', 'FORMATTED_ADDRESS', 'BROKERTITLE',
                       'MAIN_ADDRESS', 'ADMINISTRATIVE_AREA_LEVEL_2', 'ADDRESS']

ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_train_encoded = ohe.fit_transform(X_train[categorical_columns])
X_test_encoded = ohe.transform(X_test[categorical_columns])

encoded_feature_names = ohe.get_feature_names_out(categorical_columns)

X_train_encoded_d = pd.DataFrame(X_train_encoded, columns=encoded_feature_names, index=X_train.index)
X_test_encoded_d = pd.DataFrame(X_test_encoded, columns=encoded_feature_names, index=X_test.index)

X_test_encoded
```

```python
<b>4. Оформляем данные в виде класса Dataset из библиотеки torch</b>
```

```python
import torch
from torch.utils.data import Dataset

y_train_log = y_train_log.values
y_test_log = y_test_log.values

class CustomDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        x_tensor = torch.tensor(self.X[idx], dtype=torch.float32)
        y_tensor = torch.tensor(self.y[idx], dtype=torch.float32)
        return x_tensor, y_tensor

train = CustomDataset(X_train_encoded, y_train_log)
test = CustomDataset(X_test_encoded, y_test_log)
train[0]
```

<b>В Dataloader</b>

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(train, batch_size=128, shuffle=True)
test_dataloader = DataLoader(test, batch_size=128, shuffle=True)

train_features, train_labels = next(iter(train_dataloader))
train_labels = train_labels.unsqueeze(1)
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")
```

<b>5. Обучаем на тренировочном множестве</b>

<b>A. Очень простую однослойную нейросеть с оптимизатором SGD</b>

```python
import torch.nn as nn
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.linear = nn.Linear(14616, 1)

    def forward(self, x):
        return self.linear(x)
```

```python
model1 = SimpleNN()

learning_rate1 = 1e-3
batch_size = 128
epochs = 100

loss_fn = nn.MSELoss()

optimizer1 = torch.optim.SGD(model1.parameters(), lr=learning_rate1)

def train_loop1(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    total_loss = 0
    num_batches = len(dataloader)
    for batch, (X, y) in enumerate(dataloader):
        y = y.unsqueeze(1)
        pred = model(X)
        loss = loss_fn(pred, y)
        total_loss += loss.item()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if batch % 15 == 0:
            loss = loss.item()
            current = batch * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
    print(total_loss/num_batches)
    return total_loss/num_batches

def test_loop1(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss = 0

    with torch.no_grad():
        for X, y in dataloader:
            y = y.unsqueeze(1)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()

    test_loss /= num_batches
    print(f"Test Error: \n Avg loss: {test_loss:>8f} \n")
    return test_loss
```

<b>B. Нейросеть посложнее (с 1 скрытым слоем) с оптимизатором Adam</b>

```python
class OneHidLNN(nn.Module):
    def __init__(self):
        super(OneHidLNN, self).__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(14616, 528),    
            nn.ReLU(),           
            nn.Linear(528, 1),       
        )

    def forward(self, x):
        return self.linear_relu_stack(x)
```

```python
model2 = OneHidLNN()

learning_rate2 = 1e-3
batch_size = 128
epochs = 100

loss_fn = nn.MSELoss()

optimizer2 = torch.optim.Adam(model2.parameters(), lr=learning_rate2)

def train_loop2(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    total_loss = 0
    num_batches = len(dataloader)
    for batch, (X, y) in enumerate(dataloader):
        y = y.unsqueeze(1)
        pred = model(X)
        loss = loss_fn(pred, y)
        total_loss += loss.item() 
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if batch % 15 == 0:
            loss, current = loss.item(), batch * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
    print(f"Avg loss: {total_loss/num_batches:>8f}")
    return total_loss/num_batches

def test_loop2(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss = 0

    with torch.no_grad():
        for X, y in dataloader:
            y = y.unsqueeze(1)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()

    test_loss /= num_batches
    print(f"Test Error: \n Avg loss: {test_loss:>8f} \n")
    return test_loss
    
```

<b>C. Нейросеть еще сложнее (с 3+ скрытыми слоями) с оптимизатором Adam</b>

```python
class ThreeHidLNN(nn.Module):
    def __init__(self):
        super(ThreeHidLNN, self).__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(14616, 1024),
            nn.BatchNorm1d(1024), 
            nn.ReLU(),     
            nn.Dropout(0.5),
            nn.Linear(1024, 512),
            nn.BatchNorm1d(512),   
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, 256),  
            nn.BatchNorm1d(256),     
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 1)
        )

    def forward(self, x):
        return self.linear_relu_stack(x)
```

```python
model3 = ThreeHidLNN()

learning_rate3 = 0.0001
batch_size = 128
epochs3 = 200

loss_fn = nn.MSELoss()

optimizer3 = torch.optim.Adam(model3.parameters(), lr=learning_rate3)

def train_loop3(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    total_loss = 0
    num_batches = len(dataloader)
    for batch, (X, y) in enumerate(dataloader):
        y = y.unsqueeze(1)
        pred = model(X)
        loss = loss_fn(pred, y)
        total_loss += loss.item()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), batch * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
    print(f"Avg loss: {total_loss/num_batches:>8f}")
    return total_loss/num_batches

def test_loop3(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss = 0

    with torch.no_grad():
        for X, y in dataloader:
            y = y.unsqueeze(1)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()

    test_loss /= num_batches
    print(f"Test Error: \n Avg loss: {test_loss:>8f} \n")
    return test_loss
```
